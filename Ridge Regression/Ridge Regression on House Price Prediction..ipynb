# Importing the libraries.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import numpy as np
```

# Importing the dataset.

```python
data=pd.read_csv('train.csv')
data
```

# Performing EDA.

```python
data.shape
```

```python
data.head()
```

```python
# getting the null values
data.isnull().sum().sort_values(ascending = False)
```

```python
#checking how much percent of that column is missing
data.isnull().mean().sort_values(ascending = False)
```

```python
#checking correlation of SalePrice with other columns
corr_ser = data.corr().iloc[-1,:].sort_values(ascending = False)
corr_ser
```

```python
#selecting top 10 predictors
columns = corr_ser.index[:10]
columns
```

```python
df1 = data.loc[:,columns]
```

```python
df1.head()
```

```python
df1.isna().sum()
```

```python
#seperating X and y
X = df1.iloc[:,1:].values
y = df1.iloc[:,0].values
```

```python
X
```

```python
#splitting train and test values
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size= 0.2,random_state = 0)
```

# Model building.

```python

from sklearn.linear_model import Ridge
```

```python
model = Ridge(normalize=True)
```

```python

model.fit(X_train, y_train)
```

```python
#prediction
y_pred = model.predict(X_test)
```

# Model evaluation.

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```
